```python
%matplotlib inline
import requests
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.patches import Circle, Rectangle, Arc

```

```python
shot_x = [-35.93, 175.57, -76.73, 63.50, -123.88, -140.49, 187.30, -11.68, 83.99, 12.19,
           50.70, -185.56, 195.05, -15.13, -101.62, 56.47, -81.30, -175.20, 88.49, -30.13]

shot_y = [126.46, 499.77, 667.74, 192.63, 303.68, 79.23, 617.12, 251.74, 745.17, 421.21,
          358.59, 290.52, 544.82, 684.76, 123.99, 490.13, 370.36, 412.34, 278.18, 189.02]
```

```python
def draw_court(ax=None, color='black', lw=3, outer_lines=False):
    # If an axes object isn't provided to plot onto, just get current one
    if ax is None:
        ax = plt.gca()
    
    # Área de gol na parte inferior da quadra
    gol_area_line_bottom = Rectangle((-16, 120), 32, 0, linewidth=lw, color=color)
    gol_area_right_arc_bottom = Arc((16, 0), 240, 240, theta1=0, theta2=90, linewidth=lw, color=color)
    gol_area_left_arc_bottom = Arc((-16, 0), 240, 240, theta1=90, theta2=180, linewidth=lw, color=color)
    
    # Área de gol na parte superior da quadra (espelhada verticalmente)
    gol_area_line_top = Rectangle((-16, 680), 32, 0, linewidth=lw, color=color)
    gol_area_right_arc_top = Arc((16, 800), 240, 240, theta1=270, theta2=360, linewidth=lw, color=color)
    gol_area_left_arc_top = Arc((-16, 800), 240, 240, theta1=180, theta2=270, linewidth=lw, color=color)
    
    center_line = Rectangle((-200, 400), 400, 0, linewidth=lw, color=color)
    center_circle = Circle((0, 400), 60, linewidth=lw, color=color, fill=False)

    # Lista de todos os elementos da quadra, incluindo as áreas de gol
    court_elements = [
        gol_area_line_bottom, gol_area_right_arc_bottom, gol_area_left_arc_bottom,
        gol_area_line_top, gol_area_right_arc_top, gol_area_left_arc_top,center_line,center_circle
    ]
    
    if outer_lines:
        # Linha externa da quadra
        outer_lines = Rectangle((-200, 0), 400, 800, linewidth=lw, color=color, fill=False)
        court_elements.append(outer_lines)
    
    # Adiciona todos os elementos da quadra ao eixo
    for element in court_elements:
        ax.add_patch(element)
    
    ax.set_aspect('equal')
    return ax
```

```python
plt.figure(figsize=(12,11))
draw_court(outer_lines=True)
plt.scatter(shot_x, shot_y, color='red', s=50, edgecolor='black', label="Tentativas de Chute")  
plt.xlim(-300,300)
plt.ylim(-100,1000)
plt.show()
```
